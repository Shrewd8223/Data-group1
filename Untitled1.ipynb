```python
import json
import matplotlib.pyplot as plt
with open('combined_data2018.json', 'r') as f:
 data = json.load(f)
states = sorted(data['data']['states'], key=lambda x: x['registeredUsers'],reverse=True)
state_names = [state['name'].upper() for state in states]
 
registered_users = [state['registeredUsers'] for state in states]
 
plt.figure(figsize=(20,10))
plt.bar(state_names, registered_users)
plt.title('Top 10 States for Registered Users in 2018')
plt.xlabel('State'.upper())
plt.ylabel('Number of Registered Users')
plt.show()


import json
import matplotlib.pyplot as plt
with open('combined_data2019.json', 'r') as f:
 data = json.load(f)
states = sorted(data['data']['states'], key=lambda x: x['registeredUsers'],reverse=True)
state_names = [state['name'].upper() for state in states]
 
registered_users = [state['registeredUsers'] for state in states]
 
plt.figure(figsize=(20,10))
plt.bar(state_names, registered_users)
plt.title('Top 10 States for Registered Users in 2018')
plt.xlabel('State'.upper())
plt.ylabel('Number of Registered Users')
plt.show()

import json
import matplotlib.pyplot as plt
with open('combined_data2020.json', 'r') as f:
 data = json.load(f)
states = sorted(data['data']['states'], key=lambda x: x['registeredUsers'],reverse=True)
state_names = [state['name'].upper() for state in states]
 
registered_users = [state['registeredUsers'] for state in states]
 
plt.figure(figsize=(20,10))
plt.bar(state_names, registered_users)
plt.title('Top 10 States for Registered Users in 2018')
plt.xlabel('State'.upper())
plt.ylabel('Number of Registered Users')
plt.show()

import json
import matplotlib.pyplot as plt
with open('combined_data2021.json', 'r') as f:
 data = json.load(f)
states = sorted(data['data']['states'], key=lambda x: x['registeredUsers'],reverse=True)
state_names = [state['name'].upper() for state in states]
 
registered_users = [state['registeredUsers'] for state in states]
 
plt.figure(figsize=(20,10))
plt.bar(state_names, registered_users)
plt.title('Top 10 States for Registered Users in 2018')
plt.xlabel('State'.upper())
plt.ylabel('Number of Registered Users')
plt.show()

import json
import matplotlib.pyplot as plt
with open('combined_data2022.json', 'r') as f:
 data = json.load(f)
states = sorted(data['data']['states'], key=lambda x: x['registeredUsers'],reverse=True)
state_names = [state['name'].upper() for state in states]
 
registered_users = [state['registeredUsers'] for state in states]
 
plt.figure(figsize=(20,10))
plt.bar(state_names, registered_users)
plt.title('Top 10 States for Registered Users in 2018')
plt.xlabel('State'.upper())
plt.ylabel('Number of Registered Users')
plt.show()
```
